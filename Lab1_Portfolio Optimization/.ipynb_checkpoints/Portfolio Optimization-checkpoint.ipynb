```python
import numpy as np
import matplotlib.pylab as plt
import scipy.io

import warnings
warnings.filterwarnings("ignore")
```

```python
HSI = scipy.io.loadmat('./TP1/HSI50_2011_2013.mat')
NYSE = scipy.io.loadmat('./TP1/logreturn_NYSE148_20102013.mat')
SP100 = scipy.io.loadmat('./TP1/logreturn_SP100_2011_2013.mat')
SP500 = scipy.io.loadmat('./TP1/logreturn_SP500_2010_2013.mat')
```

```python
HSI = HSI['Y']
NYSE = NYSE['Y']
SP100 = SP100['Y']
SP500 = SP500['Y']
```

```python
print("The shape of HSI data:", HSI.shape)
print("The shape of NYSE data:", NYSE.shape)
print("The shape of SP100 data:", SP100.shape)
print("The shape of SP500 data:", SP500.shape)
```

```python
N_asset_HSI = np.size(HSI,0)
N_asset_NYSE = np.size(NYSE,0)
N_asset_SP100 = np.size(SP100,0)
N_asset_SP500 = np.size(SP500,0)

N_days_HSI = np.size(HSI,1)
N_days_NYSE = np.size(NYSE,1)
N_days_SP100 = np.size(SP100,1)
N_days_SP500 = np.size(SP500,1)
```

```python
w_uniform_HSI = (1/N_asset_HSI)*np.ones((N_asset_HSI,1))
w_uniform_NYSE  = (1/N_asset_NYSE)*np.ones((N_asset_NYSE,1))
w_uniform_SP100 = (1/N_asset_SP100)*np.ones((N_asset_SP100,1))
w_uniform_SP500 = (1/N_asset_SP500)*np.ones((N_asset_SP500,1))
```

```python
# xTw = lambda x,w: x.T*w
xTw = lambda x,w: x.T@w
```

```python
fig1, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(12,8))



ax1.plot(xTw(np.exp(HSI[:,N_days_HSI-30:]),w_uniform_HSI))
ax1.set_title('Volatility for HSI assets last 30 days')
ax2.plot(xTw(np.exp(NYSE[:,N_days_NYSE-30:]),w_uniform_NYSE))
ax2.set_title('Volatility for NYSE assets last 30 days')

ax3.plot(xTw(np.exp(SP100[:,N_days_SP100-30:]),w_uniform_SP100))
ax3.set_title('Volatility for SP100 assets last 30 days')

ax4.plot(xTw(np.exp(SP500[:,N_days_SP500-30:]),w_uniform_SP500))
ax4.set_title('Volatility for SP500 assets last 30 days')

# plt.savefig('volatility_last_30_days.eps')
# plt.savefig('volatility of last 30 days.png')
```

```python
print("Volatility of last 30 days HSI",np.var(xTw(np.exp(HSI[:,N_days_HSI-30:]), w_uniform_HSI)))
print("Volatility of last 30 days NYSE",np.var(xTw(np.exp(NYSE[:,N_days_NYSE-30:]), w_uniform_NYSE)))
print("Volatility of last 30 days SP100",np.var(xTw(np.exp(SP100[:,N_days_SP100-30:]), w_uniform_SP100)))
print("Volatility of last 30 days SP500",np.var(xTw(np.exp(SP500[:,N_days_SP500-30:]), w_uniform_SP500)))
```

```python
C_HSI = np.cov(np.exp(HSI[:,:N_days_HSI-30]))
C_NYSE = np.cov(np.exp(NYSE[:,:N_days_NYSE-30]))
C_SP100 = np.cov(np.exp(SP100[:,:N_days_SP100-30]))
C_SP500 = np.cov(np.exp(SP500[:,:N_days_SP500-30]))

# C_HSI = np.cov(xTw(np.exp(HSI[:,:N_days_HSI-30]), w_uniform_HSI))
# C_NYSE = np.cov(xTw(np.exp(NYSE[:,:N_days_NYSE-30]), w_uniform_NYSE))
# C_SP100 = np.cov(xTw(np.exp(SP100[:,:N_days_SP100-30]), w_uniform_SP100))
# C_SP500 = np.cov(xTw(np.exp(SP500[:,:N_days_SP500-30]), w_uniform_SP500))

print("C for HSI:", C_HSI.shape)
print("C for NYSE:", C_NYSE.shape)
print("C for SP100:", C_SP100.shape)
print("C for SP500:", C_SP500.shape)
```

```python
w_unconstraint = lambda c,n: (np.linalg.inv(c)@np.ones((n,1)))/(np.ones((n,1)).T@np.linalg.inv(c)@np.ones((n,1)))
```

```python
w_uncons_HSI = w_unconstraint(C_HSI,N_asset_HSI)
w_uncons_NYSE = w_unconstraint(C_NYSE,N_asset_NYSE)
w_uncons_SP100 = w_unconstraint(C_SP100,N_asset_SP100)
w_uncons_SP500 = w_unconstraint(C_SP500,N_asset_SP500)
```

```python
print("Volatility of day 1 to 30 days to the last with optimization HSI",np.var(xTw(np.exp(HSI[:,:N_days_HSI-30]), w_uncons_HSI)))
print("Volatility of day 1 to 30 days to the last without optimization HSI",np.var(xTw(np.exp(HSI[:,:N_days_HSI-30]), w_uniform_HSI)))
print("")

print("Volatility of day 1 to 30 days to the last with optimization NYSE",np.var(xTw(np.exp(NYSE[:,:N_days_NYSE-30]), w_uncons_NYSE)))
print("Volatility of day 1 to 30 days to the last without optimization NYSE",np.var(xTw(np.exp(NYSE[:,:N_days_NYSE-30]), w_uniform_NYSE)))

print("")
print("Volatility of day 1 to 30 days to the last with optimization SP100",np.var(xTw(np.exp(SP100[:,:N_days_SP100-30]), w_uncons_SP100)))
print("Volatility of day 1 to 30 days to the last without optimization SP100",np.var(xTw(np.exp(SP100[:,:N_days_SP100-30]), w_uniform_SP100)))

print("")
print("Volatility of day 1 to 30 days to the last with optimization SP500",np.var(xTw(np.exp(SP500[:,:N_days_SP500-30]), w_uncons_SP500)))
print("Volatility of day 1 to 30 days to the last without optimization SP500",np.var(xTw(np.exp(SP500[:,:N_days_SP500-30]), w_uniform_SP500)))
```

```python
# function for minimization f = W.T@C@W
def grad_f(w,c):
    return 2*c@w
def gradient_descent(c, step=1e-1):
    i = 0
    grad = 100
    tk = step
    w=np.random.randn(np.size(c,0),1)
    while(np.linalg.norm(grad)>1e-2):
        grad = grad_f(w,c)
        w = w-tk*grad
        #this is the projection step
        w[w<0]=0
        i+=1
        if i%5000==0:
            print(i)
    return  w
```

```python
# projected gradient descent implementation
def projection_oracle_l1 (w, l1_norm):
    # first remember signs and store them. Modify w
    signs = np.sign(w)
    w = w*signs
    
    # project this modified w onto the simplex in first orthant.
    d=len(w)
    if np.sum(w)<=l1_norm:
        return w #*signs
    for i in range(d):
        w_next = w+ 0
        w_next[w>1e-7 ] = w[w> 1e-7 ] - np.min(w[w> 1e-7 ])
        if np.sum(w_next)<=l1_norm:
            w = ((l1_norm - np.sum(w_next))*w + (np.sum(w) -
                     l1_norm)*w_next)/(np.sum(w)-np.sum(w_next))
            return w#*signs
        else:
            w=w_next
def gradient_descent2(c):
    eta= 1.0
    for l1_norm in np.arange(0, 10,0.5):
        w=np.random.randn(np.size(c,0),1)
        for i in range(1000):
            w = w - eta*grad_f(w,c)
            w = projection_oracle_l1(w, l1_norm)
    return w
```

```python
w_naive_HSI = gradient_descent(C_HSI)
w_naive_NYSE = gradient_descent(C_NYSE)
w_naive_SP100 = gradient_descent(C_SP100)
w_naive_SP500 = gradient_descent(C_SP500)
```

```python
def function (w, c):
    return np.dot(w.T, np.dot(c, w))
def minimization (w, c, m):
    return np.dot(w.T, np.dot(c, w)) - (1/m)*np.log(np.sum(w))

def line_search(c, w, t, df, dw, m, alpha = 0.1, beta = 0.7):
    '''
    Backtracking line search.
    '''
    if minimization (w + t*dw, c, m).all() <= (minimization (w, c, m) + alpha*t*np.dot(df.T, dw)).all() or ((w+t*dw).all()>0):
        return w + t*dw
    else:
        return line_search(c, w, beta*t, df, dw, m)

def centering_step(c, t, w0, eps):
    df = t*(2*np.dot(c, w0)) + sum([1/(w0)])
    d2f = 2*t*c + sum([1/(w0**2)])
    dw = -1*np.dot(np.linalg.inv(d2f), df)
    l2 = np.dot(df.T, np.dot(np.linalg.inv(d2f), df))
    if l2/2 <= eps:
        return w0
    w1 = line_search(c, w0, t=1, df=df, dw=dw, m=t)
    return centering_step(c, t, w1, eps)
               
def barr_method(c, w0, eps, mu, t=1):
    w_center = centering_step(c, t, w0, eps)
    if 1/t < eps:
        return w_center
    else:
        t = mu*t
        return barr_method(c, w0, eps, t, mu)
               
def w_constraint(n, w):
    P = np.identity(n) - (1/n)*np.dot(np.ones((n,1)),np.ones((n,1)).T)
    w0 = np.ones_like(w)
    w0 [1:] = w[1:]
    w1 = np.dot(P, w0) + (1/n)*np.ones((n,1))
    return w1
    
def w_constraint2(n, w):
    P = np.identity(n) - (1/n)*np.dot(np.ones((n,1)),np.ones((n,1)).T)
    w0 = np.ones_like(w)/n
    w0 [1:] = w[1:]
    w1 = np.dot(P, w) + (1/n)*np.ones((n,1))
    return w1
```

```python
eps = 10e-6
w0_HSI = np.random.randn(N_asset_HSI,1)
w0_NYSE = np.random.randn(N_asset_NYSE,1)
w0_SP100 = np.random.randn(N_asset_SP100,1)
w0_SP500 = np.random.randn(N_asset_SP500,1)
mu_list = [2, 5, 10, 15, 20, 25, 30, 40, 50, 70, 85, 100, 130, 170, 200, 250]
for mu in mu_list:
    w_const_HSI= barr_method(C_HSI, w0_HSI, eps, mu)
    w_const_NYSE= barr_method(C_NYSE, w0_NYSE, eps, mu)
    w_const_SP100= barr_method(C_SP100, w0_SP100, eps, mu)
    w_const_SP500= barr_method(C_SP500, w0_SP500, eps, mu)
    
w_const1_HSI = w_constraint(N_asset_HSI, w_const_HSI)
w_const1_NYSE = w_constraint(N_asset_NYSE, w_const_NYSE)
w_const1_SP100 = w_constraint(N_asset_SP100, w_const_SP100)
w_const1_SP500 = w_constraint(N_asset_SP500, w_const_SP500)

w_const2_HSI = w_constraint2(N_asset_HSI, w_const_HSI)
w_const2_NYSE = w_constraint2(N_asset_NYSE, w_const_NYSE)
w_const2_SP100 = w_constraint2(N_asset_SP100, w_const_SP100)
w_const2_SP500 = w_constraint2(N_asset_SP500, w_const_SP500)

plt.figure(figsize = (12,6))
plt.plot(xTw(HSI[:,:N_asset_HSI-30], w_uniform_HSI), "red", label="uniform allocation")
plt.plot(xTw(HSI[:,:N_asset_HSI-30], w_uncons_HSI), "blue", label = "unconstrained optimization")
plt.plot(xTw(HSI[:,:N_asset_HSI-30], w_const1_HSI), "black", label = "constrained optimization")
plt.title("HSI data set")
plt.legend()

plt.figure(figsize = (12,6))
plt.plot(xTw(NYSE[:,:N_asset_NYSE-30], w_uniform_NYSE), "red", label="uniform allocation")
plt.plot(xTw(NYSE[:,:N_asset_NYSE-30], w_uncons_NYSE), "blue", label = "unconstrained optimization")
plt.plot(xTw(NYSE[:,:N_asset_NYSE-30], w_const1_NYSE), "black", label = "constrained optimization")
plt.title("NYSE data set")
plt.legend()

plt.figure(figsize = (12,6))
plt.plot(xTw(SP100[:,:N_asset_SP100-30], w_uniform_SP100), "red", label="uniform allocation")
plt.plot(xTw(SP100[:,:N_asset_SP100-30], w_uncons_SP100), "blue", label = "unconstrained optimization")
plt.plot(xTw(SP100[:,:N_asset_SP100-30], w_const1_SP100), "black", label = "constrained optimization")
plt.title("SP100 data set")
plt.legend()

plt.figure(figsize = (12,6))
plt.plot(xTw(SP500[:,:N_asset_SP500-30], w_uniform_SP500), "red", label="uniform allocation")
plt.plot(xTw(SP500[:,:N_asset_SP500-30], w_uncons_SP500), "blue", label = "unconstrained optimization")
plt.plot(xTw(SP500[:,:N_asset_SP500-30], w_const1_SP500), "black", label = "constrained optimization")
plt.title("SP500 data set")
plt.legend()

print("Volatility of last 30 days for HSI by constraint method",np.var(xTw(np.exp(HSI[:,N_days_HSI-30:]), w_const2_HSI)))
print("Volatility of last 30 days for HSI unconstrained method",np.var(xTw(np.exp(HSI[:,N_days_HSI-30:]), w_uncons_HSI)))
print("Volatility of last 30 days for HSI uniform method",np.var(xTw(np.exp(HSI[:,N_days_HSI-30:]), w_uniform_HSI)))
print("")

print("Volatility of last 30 days for NYSE by constraint method",np.var(xTw(np.exp(NYSE[:,N_days_NYSE-30:]), w_const2_NYSE)))
print("Volatility of last 30 days for NYSE unconstrained method",np.var(xTw(np.exp(NYSE[:,N_days_NYSE-30:]), w_uncons_NYSE)))
print("Volatility of last 30 days for NYSE uniform method",np.var(xTw(np.exp(NYSE[:,N_days_NYSE-30:]), w_uniform_NYSE)))
print("")

print("Volatility of last 30 days for SP100 by constraint method",np.var(xTw(np.exp(SP100[:,N_days_SP100-30:]), w_const2_SP100)))
print("Volatility of last 30 days for SP100 unconstrained method",np.var(xTw(np.exp(SP100[:,N_days_SP100-30:]), w_uncons_SP100)))
print("Volatility of last 30 days for SP100 uniform method",np.var(xTw(np.exp(SP100[:,N_days_SP100-30:]), w_uniform_SP100)))
print("")

print("Volatility of last 30 days for SP500 by constraint method",np.var(xTw(np.exp(SP500[:,N_days_SP500-30:]), w_const2_SP500)))
print("Volatility of last 30 days for SP500 unconstrained method",np.var(xTw(np.exp(SP500[:,N_days_SP500-30:]), w_uncons_SP500)))
print("Volatility of last 30 days for SP500 uniform method",np.var(xTw(np.exp(SP500[:,N_days_SP500-30:]), w_uniform_SP500)))
```
